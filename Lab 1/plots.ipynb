#### Plot of $f(x) = 10x^4 + 3x^3 - 30x^2 + 10x$

```python
import matplotlib.pyplot as plt
from config import (
    F, X, Y_F
)

plt.plot(X, Y_F, label=str(F))

plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.grid(True, linestyle='--', alpha=0.6)
plt.ylim(-50, 30)
plt.title("Plot of f(x)")
plt.xlabel("x")
plt.ylabel("f(x)")


plt.show()
```

#### Plot of $g(x_1, x_2) = (x_1 -2)^4 + (x_2 + 3)^4 + 2(x_1 -2)^2 (x_2 + 3)^2$

```python
from config import (
    X1, X2, Y_G
)


fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')


ax.plot_surface(X1, X2, Y_G, cmap='viridis', alpha=0.8)


ax.contour3D(X1, X2, Y_G, levels=30, cmap='magma')


ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_zlabel('$g(x_1, x_2)$')


plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.2)


plt.show()
```

```python
plt.figure(figsize=(10, 7))
contour = plt.contourf(X1, X2, Y_G, levels=30, cmap='viridis')
plt.colorbar(contour)
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.title('Contour map of $g(x_1, x_2)$')
plt.show()
```
