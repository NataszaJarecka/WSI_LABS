#### Plot of $f(x) = 10x^4 + 3x^3 - 30x^2 + 10x$

```python
import numpy as np
import matplotlib.pyplot as plt

f_coeffs = [10, 3, -30, 10, 0]
f= np.poly1d(f_coeffs)
x = np.linspace(-3, 3, 400)
y = f(x)

plt.plot(x, y, label=str(f))

plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.grid(True, linestyle='--', alpha=0.6)
plt.ylim(-50, 30)
plt.title("Plot of f(x)")
plt.xlabel("x")
plt.ylabel("f(x)")


plt.show()
```

#### Plot of $g(x_1, x_2) = (x_1 -2)^4 + (x_2 + 3)^4 + 2(x_1 -2)^2 (x_2 + 3)^2$

```python
x1 = np.linspace(-50, 50, 50)
x2 = np.linspace(-50, 50, 50)

X1, X2 = np.meshgrid(x1, x2)
g = (X1-2)**4 + (X2+3)**4 + 2 * ((X1-2)**2)*((X2+3)**2)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X1, X2, g, cmap='viridis', alpha=0.8)
ax.contour3D(X1, X2, g, levels = 30, cmap='magma')


ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_zlabel('$g(x_1, x_2)$')

plt.show()
```

```python
plt.figure(figsize=(10, 7))
contour = plt.contourf(X1, X2, g, levels=30, cmap='viridis')
plt.colorbar(contour)
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.title('Contour map of $g(x_1, x_2)$')
plt.show()
```
