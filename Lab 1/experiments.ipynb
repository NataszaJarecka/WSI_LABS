## Searching for minimum of $f(x)$


```python
from gradient_descent_algorythm import gradient_descent
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

x = sp.symbols('x')

f = lambda x: 10*x**4 + 3*x**3 - 30*x**2 + 10*x
variables = [x]
step = 0.01
start = [-2]
presicion = 0.000001
iter = 10000

point, i, all_points = gradient_descent(f, variables, step, start, presicion, iter)

print(f'Minimum in x = {point}, found in {i} iterations')

# visualisation
f_coeffs = [10, 3, -30, 10, 0]
f_poly = np.poly1d(f_coeffs)
x1 = np.linspace(-2.5, 0, 400)
y1 = f_poly(x1)
y2 = f_poly(all_points)


plt.plot(x1, y1, label=str(f_poly))
plt.plot(all_points, y2, marker='o', linestyle='-', color='r', label="Path of gradient", markersize=4)

plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.grid(True, linestyle='--', alpha=0.6)
plt.ylim(-50, 30)
plt.title("Plot of f(x)")
plt.xlabel("x")
plt.ylabel("f(x)")


plt.show()
```

## Searching for minimum of $g(x_1, x_2)$


```python
x1, x2 = sp.symbols('x1 x2')

g = lambda x1, x2:(x1-2)**4 + (x2 + 3)**4 + 2*((x1 - 2)**2)*((x2 + 3)**2)
variables = [x1, x2]
step = 0.03
start = [0, 0]
presicion = 0.000001
iter = 10000

point, i = gradient_descent(g, variables, step, start, presicion, iter)

print(f'Minimum in x = {point}, found in {i} iterations')

```
