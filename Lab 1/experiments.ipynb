## Searching for minimum of $f(x)$


```python
from gradient_descent_algorythm import gradient_descent
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np
from config import (
    F, X, Y_F, X1, X2, Y_G, F_VARIABLES, F_ITER, F_PRECISION, G_VARIABLES, G_PRECISION, G_ITER, G, F_GLOBAL_MINIMUM, REPEATS, G_GLOBAL_MINIMUM
)

step = 0.005
start = [-2]

point, i, all_points = gradient_descent(F, F_VARIABLES, step, start, F_PRECISION, F_ITER)

print(f'Minimum in x = {point}, found in {i} iterations')

# visualisation

all_points = np.array(all_points)

y_points = F(all_points)

plt.plot(X, Y_F, label=str(F))

plt.plot(all_points, y_points, marker='o', linestyle='-', color='r', label="Path of gradient", markersize=4)


plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.grid(True, linestyle='--', alpha=0.6)
plt.ylim(-50, 30)
plt.xlim(-2.5, 0.5)
plt.title("Plot of f(x)")
plt.xlabel("x")
plt.ylabel("f(x)")


plt.show()
```

Checking the percentage rate of success of algorithm

```python

step = 0.0001

minima_found = 0


for k in range(REPEATS):

    a = random.uniform(-25, 25)
    b = random.uniform(-25, 25)

    start = [a, b]

    point, i, all_points = gradient_descent(G, G_VARIABLES, step, start, G_PRECISION, G_ITER)

    if np.linalg.norm(point - G_GLOBAL_MINIMUM) <= G_PRECISION:
        minima_found += 1

print(F'Global minimum found in {(minima_found/REPEATS)*100} % of cases with precision {G_PRECISION}')

minima_found = 0

small_precision = 0.5

for k in range(REPEATS):

    a = random.uniform(-25, 25)
    b = random.uniform(-25, 25)

    start = [a, b]

    point, i, all_points = gradient_descent(G, G_VARIABLES, step, start, small_precision, G_ITER)

    if np.linalg.norm(point - G_GLOBAL_MINIMUM) <= small_precision:
        minima_found += 1

print(F'Global minimum found in {(minima_found/REPEATS)*100} % of cases with precision {small_precision}')



```
